## **Step 2:** Using BeautifulSoup Scraping List of Postal Codes of Given Wikipedia Page 

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
extracting_data = requests.get(url)
wiki_data = BeautifulSoup(extracting_data.content, 'html.parser')
```

## **Step 3:** Converting content of PostalCode HTML table as dataframe

```python
table = wiki_data.find('tbody')
rows = table.select('tr')
row = [r.get_text() for r in rows]
```

```python
df = pd.DataFrame(row)
df1 = df[0].str.split('\n', expand=True)
df2 = df1.rename(columns=df1.iloc[0])
df3 = df2.drop(df2.index[0])
df3.head(12)
```

## **Step 4:** Data Cleaning | Drop None rows of df and row which contains 'Not assigned' value 

```python
df4 = df3[df3.Borough != 'Not assigned']
df4.head(12)
```

```python
df5 = df4.groupby(['Postal Code', 'Borough'], sort = False).agg(','.join)
df6 = df5.reset_index()
df6.head(12)
```
